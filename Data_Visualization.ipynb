```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_data.csv')
```

**DATA VISUALTION**

```python
sns.set_style('whitegrid')
```

*What is the most frequent genre of movies released on Netflix*

```python
df['Genre'].describe()
```

```python
sns.catplot(y='Genre',data=df,kind='count',order=df['Genre'].value_counts().index)
plt.title('Frequency of movies Genre relead on Netflix')
plt.tight_layout()
plt.show()
```

*Which has highest Average Vote*

```python
df['Vote_Average'].describe()
```

```python
sns.catplot(y='Vote_Average',data=df,kind='count',order=df['Vote_Average'].value_counts().index)
plt.title('Highest Average Vote')
plt.tight_layout()
plt.show()
```

*What movie got the highest Popularity and what's its genre*

```python
df[df['Popularity']==df['Popularity'].max()]
```

*What movie got the lowest Popularity and what's its genre*

```python
df[df['Popularity']==df['Popularity'].min()]
```

*Which year has the most filmed movie*

```python
plt.figure(figsize=(10, 6))  # Width=12 inches, Height=6 inches
df['Release_Date'].hist()
plt.title('Release Date Column Distribution')
plt.xlabel('Release Date')
plt.ylabel('Number of movies Released')
plt.tight_layout()
plt.show()
```
